```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(h):
    return 1/(1 + np.exp(-h))
```

```python
x = np.linspace(-10,10,100)
y=sigmoid(x)
```

```python
#Derivitive of the sigmoid function is sigmoid(x) - sigmoid(x)^2 so,

z = sigmoid(x) - sigmoid(x) * sigmoid(x)
```

```python
plt.figure()
plt.plot(y)
plt.plot(z)
```

The limit as x approaches infinity is 1.
